```python
import numpy as np
import pandas as pd
import networkx as nx
```

```python
P = nx.Graph()
for x in range(1, 62):
    P.add_node(x)
```

```python
file = open('pedestrian_data', 'r')
for k in range(1, 105):
    x = file.readline()
    y = x.split(',')
    s, t, w = int(y[0]), int(y[1]), float(y[2])
    P.add_edge( s, t , weight=w)
```

```python
ppr = nx.pagerank(P)
```

```python
ppr_list = []
for node in ppr.keys():
    item = (node, ppr[node])
    ppr_list.append( item )
sorted(ppr_list, key=lambda x: x[1], reverse=True)
```

```python
V = nx.Graph()
for x in range(1, 62):
    V.add_node(x)
```

```python
file = open('vehicle_data', 'r')
for k in range(1, 105):
    x = file.readline()
    y = x.split(',')
    s, t, w = int(y[0]), int(y[1]), float(y[2])
    V.add_edge( s, t , weight=w)
```

```python
vpr = nx.pagerank(V)
```

```python
vpr_list = []
for node in vpr.keys():
    item = (node, vpr[node])
    vpr_list.append( item )
sorted(vpr_list, key=lambda x: x[1], reverse=True)
```

```python
car = nx.pagerank(V)
car_list = []
for node in car.keys():
    item = (node, car[node])
    car_list.append( item )
sorted(car_list, key=lambda x: x[0], reverse=True)

ppl = nx.pagerank(P)
ppl_list = []
for node in ppl.keys():
    item = (node, ppl[node])
    ppl_list.append( item )
sorted(ppl_list, key=lambda x: x[0], reverse=True)


for node in car.keys():
    item = (node, ppl[node] - car[node])
    car_list.append( item )
sorted(car_list, key=lambda x: x[1], reverse=True)
```

```python
nx.write_graphml(P, "ppl.graphml")
nx.write_graphml(V, "car.graphml")
```
